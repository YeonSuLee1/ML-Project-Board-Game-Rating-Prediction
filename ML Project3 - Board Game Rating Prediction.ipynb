# Data Visualization

```python
plt.hist(games["average_rating"])
```

```python
games = games[games["users_rated"] > 0]
games = games.dropna(axis=0)

plt.hist(games["average_rating"])
plt.show()
```

```python
# Making correlations
corr_matrix = games.corr()
corr_matrix['average_rating'].sort_values(ascending=False)
```

```python
fig = plt.figure(figsize = (20, 16))
sns.heatmap(corr_matrix,annot= True)
plt.show()
```

```python
y = games["average_rating"]
```

```python
games = games.drop(["bayes_average_rating", "average_rating", "type", "name", "id"], axis=1)
```

```python
games.head()
```

# Train Test Spliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(games, y , random_state = 0)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
y_pred = linear_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

# Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor(n_estimators=100, min_samples_leaf=10, random_state = 0)
random_forest.fit(X_train, y_train)
```

```python
y_predict = random_forest.predict(X_test)
mean_squared_error(y_predict, y_test)
```
